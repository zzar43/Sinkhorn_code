```julia
function unbalanced_sinkhorn_1d_signal1(a, b, M, reg, reg_m; reg_s=1e-1, iterMax=100, verbose=false)
    Pap = proj_p(a)
    Pan = proj_n(a)
    Pbp = proj_p(b)
    Pbn = proj_n(b)

    ap = Pap * a .+ reg_s
    an = Pan * a .+ reg_s
    bp = Pbp * b .+ reg_s
    bn = Pbn * b .+ reg_s
    
#     ap = Pap * a .+ reg_s
#     an = Pan * a .+ reg_s
#     bp = Pbp * b .+ reg_s
#     bn = Pbn * b .+ reg_s
#     ap = ap .+ reg_s*maximum(ap)
#     an = an .+ reg_s*maximum(an)
#     bp = bp .+ reg_s*maximum(bp)
#     bn = bn .+ reg_s*maximum(bn)

    if maximum(ap) == 0
        ap = ones(length(a)) ./ length(a)
    end
    if maximum(an) == 0
        an = ones(length(a)) ./ length(a)
    end
    if maximum(bp) == 0
        bp = ones(length(b)) ./ length(b)
    end
    if maximum(bn) == 0
        bn = ones(length(b)) ./ length(b)
    end

    Tp, gp, dp = unbalanced_sinkhorn_1d(ap, bp, M, reg, reg_m; iterMax=iterMax, verbose=verbose)
    Tn, gn, dn = unbalanced_sinkhorn_1d(an, bn, M, reg, reg_m; iterMax=iterMax, verbose=verbose)
    T = Tp - Tn
    # grad = Pap*gp + Pan*gn
#     grad = Pap*gp + Pan*gn
    grad = gp - gn
    dist = dp + dn

    return T, grad, dist
end
```

```julia
reg = 1e-3
reg_m = 1e2
g = ricker_func(x, 0.5, 0.05)

misfit_shift = 0.3:0.02:0.7
misfit_amp = 0.5:0.01:1.5

value_shift = zeros(length(misfit_shift))
value_amp = zeros(length(misfit_amp))

for i = 1:length(misfit_shift)
    f = ricker_func(x, misfit_shift[i], 0.05)
    TT, gg, value_shift[i] = unbalanced_sinkhorn_1d_signal1(f, g, M, reg, reg_m; reg_s=1e-2, iterMax=100, verbose=true)
end

for i = 1:length(misfit_amp)
    f = misfit_amp[i] .* ricker_func(x, 0.3, 0.05)
    TT, gg, value_amp[i] = unbalanced_sinkhorn_1d_signal1(f, g, M, reg, reg_m; reg_s=1e-2, iterMax=100, verbose=true)
end

subplot(221); plot(x,f,x,g); title("f and g"); legend(["f", "g"])
subplot(222); plot(misfit_shift, value_shift); title("convex w.r.t. shift (Not stable when reg small)")
subplot(223); plot(misfit_amp, value_amp); title("convex w.r.t. amplitude")
tight_layout()
```
